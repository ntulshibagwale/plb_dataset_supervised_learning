## Compute FFT for All Examples at Each Angle

```python
# Compute FFTs
low_pass = 0             # [Hz] ; low frequency cutoff
high_pass = 800*10**3    # [Hz] ; high frequency cutoff\
dt = 10**-7              # [seconds] ; sample period / time between samples
fft_units = 1000 

# 20 deg
fft_20deg = []
for idx,wave in enumerate(waves_20deg):
    w,z = fft(dt, wave, low_pass, high_pass)
    #z = z / (max(z)) # normalize
    fft_20deg.append(z)
fft_20deg = np.array(fft_20deg)


# 22 deg
fft_22deg = []
for idx,wave in enumerate(waves_22deg):
    w,z = fft(dt, wave, low_pass, high_pass)
    #z = z / (max(z)) # normalize
    fft_22deg.append(z)
fft_22deg = np.array(fft_22deg)

# 26 deg
fft_26deg = []
for idx,wave in enumerate(waves_26deg):
    w,z = fft(dt, wave, low_pass, high_pass)
    #z = z / (max(z)) # normalize
    fft_26deg.append(z)
fft_26deg = np.array(fft_26deg)

# 30 deg
fft_30deg = []
for idx,wave in enumerate(waves_30deg):
    w,z = fft(dt, wave, low_pass, high_pass)
    #z = z / (max(z)) # normalize
    fft_30deg.append(z)
fft_30deg = np.array(fft_30deg)

# 40 deg
fft_40deg = []
for idx,wave in enumerate(waves_40deg):
    w,z = fft(dt, wave, low_pass, high_pass)
    #z = z / (max(z)) # normalize
    fft_40deg.append(z)
fft_40deg = np.array(fft_40deg)

w = w/fft_units; # khz
print(f"# of discrete freq points for each ex: {len(w)}") # same for each wave, since freq range & dt constant
```

```python
from scipy import stats as st # t distribution
from ae_functions import *

fig,spec2 = create_figure('Frequency Spectra and Confidence Intervals for PLB at 4 Angles 4 mm Length',columns=2,rows=2,
                          width=21,height=15,default_font_size=18\
                         ,tick_font_size=20,legend_font_size=16,axes_font_size=20,\
                          title_font_size=20)
# 20 degrees
N_20deg = len(fft_20deg) # number of samples
mean_fft_20deg = np.mean(fft_20deg, axis=0) # average each column over all examples
std_fft_20deg = np.std(fft_20deg, axis=0)   # standard deviation
t95_20deg=st.t.ppf(1-0.025,N_20deg-1)
ci_20deg=t95_20deg*std_fft_20deg
ci_mean_20deg=t95_20deg*std_fft_20deg/np.sqrt(N_20deg)
upper_fft_20deg = mean_fft_20deg + ci_20deg
lower_fft_20deg = mean_fft_20deg - ci_20deg
ax_20deg = fig.add_subplot(spec2[0,0])
ax_20deg.plot(w,mean_fft_20deg,'-',label = 'Averaged FFT',color='navy')
ax_20deg.plot(w,mean_fft_20deg+ci_20deg,'--',color='orangered',linewidth=0.5,label='+/- $ \sigma $ (68% Prob)')
ax_20deg.plot(w,mean_fft_20deg-ci_20deg,'--',color='orangered',linewidth=0.5)
ax_20deg.plot(w,mean_fft_20deg+ci_mean_20deg,'--',color='red',linewidth=0.5,label='95% Confidence Interval for Mean')
ax_20deg.plot(w,mean_fft_20deg-ci_mean_20deg,'--',color='red',linewidth=0.5)
ax_20deg.legend()
ax_20deg.set_xlim([low_pass/fft_units,high_pass/fft_units])
ax_20deg.set_xlabel('Frequency (kHz)')
ax_20deg.set_ylabel('Normalized Amplitude')
plt.title('20 deg')
plt.grid()

# 26 degrees
N_26deg = len(fft_26deg) # number of samples
mean_fft_26deg = np.mean(fft_26deg, axis=0) # average each column over all examples
std_fft_26deg = np.std(fft_26deg, axis=0)   # standard deviation
t95_26deg=st.t.ppf(1-0.025,N_26deg-1)
ci_26deg=t95_26deg*std_fft_26deg
ci_mean_26deg=t95_20deg*std_fft_26deg/np.sqrt(N_26deg)
upper_fft_26deg = mean_fft_26deg + ci_26deg
lower_fft_26deg = mean_fft_26deg - ci_26deg
ax_26deg = fig.add_subplot(spec2[0,1])
ax_26deg.plot(w,mean_fft_26deg,'-',label = 'Averaged FFT',color='navy')
ax_26deg.plot(w,mean_fft_26deg+ci_26deg,'--',color='orangered',linewidth=0.5,label='+/- $ \sigma $ (68% Prob)')
ax_26deg.plot(w,mean_fft_26deg-ci_26deg,'--',color='orangered',linewidth=0.5)
ax_26deg.plot(w,mean_fft_26deg+ci_mean_26deg,'--',color='red',linewidth=0.5,label='95% Confidence Interval for Mean')
ax_26deg.plot(w,mean_fft_26deg-ci_mean_26deg,'--',color='red',linewidth=0.5)
ax_26deg.legend()
ax_26deg.set_xlim([low_pass/fft_units,high_pass/fft_units])
ax_26deg.set_xlabel('Frequency (kHz)')
ax_26deg.set_ylabel('Normalized Amplitude')
plt.title('26 deg')
plt.grid()

# 30 degrees
N_30deg = len(fft_30deg) # number of samples
mean_fft_30deg = np.mean(fft_30deg, axis=0) # average each column over all examples
std_fft_30deg = np.std(fft_30deg, axis=0)   # standard deviation
t95_30deg=st.t.ppf(1-0.025,N_30deg-1)
ci_30deg=t95_30deg*std_fft_30deg
ci_mean_30deg=t95_30deg*std_fft_30deg/np.sqrt(N_30deg)
upper_fft_30deg = mean_fft_30deg + ci_30deg
lower_fft_30deg = mean_fft_30deg - ci_30deg
ax_30deg = fig.add_subplot(spec2[1,0])
ax_30deg.plot(w,mean_fft_30deg,'-',label = 'Averaged FFT',color='navy')
ax_30deg.plot(w,mean_fft_30deg+ci_30deg,'--',color='orangered',linewidth=0.5,label='+/- $ \sigma $ (68% Prob)')
ax_30deg.plot(w,mean_fft_30deg-ci_30deg,'--',color='orangered',linewidth=0.5)
ax_30deg.plot(w,mean_fft_30deg+ci_mean_30deg,'--',color='red',linewidth=0.5,label='95% Confidence Interval for Mean')
ax_30deg.plot(w,mean_fft_30deg-ci_mean_30deg,'--',color='red',linewidth=0.5)
ax_30deg.legend()
ax_30deg.set_xlim([low_pass/fft_units,high_pass/fft_units])
ax_30deg.set_xlabel('Frequency (kHz)')
ax_30deg.set_ylabel('Normalized Amplitude')
plt.title('30 deg')
plt.grid()

# 40 degrees
N_40deg = len(fft_40deg) # number of samples
mean_fft_40deg = np.mean(fft_40deg, axis=0) # average each column over all examples
std_fft_40deg = np.std(fft_40deg, axis=0)   # standard deviation
t95_40deg=st.t.ppf(1-0.025,N_40deg-1)
ci_40deg=t95_40deg*std_fft_40deg
ci_mean_40deg=t95_40deg*std_fft_40deg/np.sqrt(N_40deg)
upper_fft_40deg = mean_fft_40deg + ci_40deg
lower_fft_40deg = mean_fft_40deg - ci_40deg
ax_40deg = fig.add_subplot(spec2[1,1])
ax_40deg.plot(w,mean_fft_40deg,'-',label = 'Averaged FFT',color='navy')
ax_40deg.plot(w,mean_fft_40deg+ci_40deg,'--',color='orangered',linewidth=0.5,label='+/- $ \sigma $ (68% Prob)')
ax_40deg.plot(w,mean_fft_40deg-ci_40deg,'--',color='orangered',linewidth=0.5)
ax_40deg.plot(w,mean_fft_40deg+ci_mean_40deg,'--',color='red',linewidth=0.5,label='95% Confidence Interval for Mean')
ax_40deg.plot(w,mean_fft_40deg-ci_mean_40deg,'--',color='red',linewidth=0.5)
ax_40deg.legend()
ax_40deg.set_xlim([low_pass/fft_units,high_pass/fft_units])
ax_40deg.set_xlabel('Frequency (kHz)')
ax_40deg.set_ylabel('Normalized Amplitude')
plt.title('40 deg')
plt.grid()

plt.show()
```
